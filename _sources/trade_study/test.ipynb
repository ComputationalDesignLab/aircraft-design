```python
import nbimporter
import numpy as np
import matplotlib.pyplot as plt
from mtow import compute_takeoff_weight

num_pts = 15
wing_loading = np.linspace(20, 40, num_pts) # lbs/sq ft
power_loading = np.linspace(5, 15, num_pts) # lbs/hp

A = 8

X, Y = np.meshgrid(wing_loading, power_loading)

mtow = np.zeros_like(X)

for i in range(num_pts):

    for j in range(num_pts):

        WS = X[i,j]
        WP = Y[i,j]

        mtow[i,j], _ = compute_takeoff_weight(WS, WP, A) # lbs

levels = np.linspace(4000, 9000, 11, dtype=int)

fs = 11

fig, ax = plt.subplots()

cp = ax.contour(X, Y, mtow, levels=levels, colors="k")

ax.clabel(cp, cp.levels, fontsize=fs)
```
