## AEO

```python
import numpy as np

A = 8
e = 0.81 - 0.05 # in TO conditions
mu = 0.04 # Raymer Table 17.1
W = 0.99 * 5374 # lbs
CD0 = 0.03363 + 0.024011 # TO conditions
CL = 0.7 * 1.8
K = 1/np.pi/A/e
L_by_D = CL / ( CD0 + K*CL**2 )
rho = 0.00237717 # slugs/cu ft, sea-level
S = 134 # ft^2

Vstall = (W/S * 2 / rho /CL)**0.5

P = 298 #hp
prop_eff = 0.8
Vclimb = 1.2 * Vstall

# Thrust
T = 550 * P * prop_eff / Vclimb
T = 0.92 * T # installation loss
T_transition = 2 * T # lbs

# Climb angle
climb_gamma = np.arcsin( T_transition/W - 1/L_by_D ) * 180 / np.pi # deg

print(climb_gamma)

print(np.tan(climb_gamma * np.pi/180) * 100)
```

## OEI

```python
import numpy as np

A = 8
e = 0.81 # clean conditions
W = 0.95 * 5374 # lbs
CD0 = 0.03363 # clean conditions
CL = 0.7 * 1.44
K = 1/np.pi/A/e
L_by_D = CL / ( CD0 + K*CL**2 )
rho = 0.00204834 # slugs/cu ft, sea-level
S = 134 # ft^2

Vstall = (W/S * 2 / rho /CL)**0.5

P = 298 #hp
prop_eff = 0.8
Vclimb = 1.2 * Vstall

# Propeller area, sq ft
dia = 5.9 # ft
Aprop = np.pi * dia**2 / 4
q = 0.5 * rho * Vclimb**2

T = 550 * P * prop_eff / Vclimb
T = 0.92 * T # installation loss
D = 0.08 * T + 0.1*Aprop*q # drag of inoperative engine

print(T,D)

T_transition = T - D # lbs

print(T_transition/W, 1/L_by_D)

# Climb angle
climb_gamma = np.arcsin( T_transition/W - 1/L_by_D ) * 180 / np.pi # deg

print(climb_gamma)

print(np.tan(climb_gamma * np.pi/180) * 100)
```
