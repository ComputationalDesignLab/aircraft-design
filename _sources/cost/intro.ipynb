# Cost analysis

This section demonstrates how to perform cost analysis for the example airplane. The cost is an important measure of merit apart from maximum takeoff weight and other performance metrics. The cost estimation is largely based on statistical relations developed using previous aircraft. This demonstration uses cost estimation relations from Chapter 2 of Gudmundsson's book. It consists of the Eastlake cost model which is a modified version of DAPCA IV model for GA airplanes. Refer to lecture notes, Gundmundsson Chapter 2, and Raymer Chapter 18 for more details about cost estimation. Following topics are covered in this section:

1. Unit cost
2. Operating cost
