# Cost analysis

This section demonstrates how to perform cost analysis for the example airplane. The cost is an important measure of merit apart from maximum takeoff weight and other performance metrics. The cost estimation is largely based on statistical relations based on previous aircraft. This demonstration uses cost estimation relations from Chapter 2 Gudmundsson. It consists of DAPCA IV model with modifications to account for GA airplanes. Refer to lecture notes, Gundmundsson Chapter 2, and Raymer Chapter 18 for more details about cost estimation.

The cost for an airplane can be broadly classified into:

1. Research, developement, test and evaluation (RDTE) cost,
2. Production cost, and
3. Operating cost

The first two costs can be combined together to yield total program cost. These costs are computed in this section, along with some additional analysis.
